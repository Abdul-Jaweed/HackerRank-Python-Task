# Q4) Nested Lists

```python
students_grade = []
for i in range(int(input())):
    name = input("Enter the Student Name : ")
    score = float(input("Enter the Student Grade : "))
    students_grade.append([name, score])
    sorted_scores = sorted(list(set([x[1] for x in students_grade ])))
    second_scores = sorted_scores[1]
    
    low_final_list = []
    for student in students_grade:
        if second_scores == student[1]:
            low_final_list.append(student[0])
    for student in sorted(low_final_list):
        print(student)
```

# Q17) Check Strick Superset

```python
def isstrictsuperset(a,b):
    # true if a is a strict superset of b
    return b.issubset(a) and not(a.issubset(b))

a = set(int(x) for x in input().split(' '))
n = int(input())
res = True

for _ in range(n):
    b = set(int(x) for x in input().split(' '))
    res &= isstrictsuperset(a,b)
    
print(res)
```
